Part 1: Prompt Engineering for Mass Personalization

Prompt 1: Targeting CTO of a D2C Brand (AIDCA + Cialdini Principles)
Goal: High-converting email for CTOs struggling with customer retention.

```python
# @title ### **AIDCA Email Generator with Cialdini Principles**

def generate_email(prompt):
    # In production, this would use OpenAI API
    # For simulation, we'll use a simplified version
    response = f"""
    **Generated Email:**
    {prompt}

    **Performance Prediction:**
    - Expected Open Rate: 42-48%
    - Expected CTR: 8-12%
    - Conversion Lift: 3-5x vs generic template
    """
    return response

# Prompt 1: CTO of D2C Brand
cto_prompt = """
Generate a high-conversion email for D2C CTOs using AIDCA + Cialdini's Authority & Scarcity:

**Attention:** "85% of D2C brands lose revenue to cart abandonment—but yours doesn't have to."
**Interest:** "See how [Similar Brand] increased retention by 30% with our AI workflow (case study attached)."
**Desire:** "Only 5 spots remain for our Q3 onboarding program."
**Conviction:** "Featured in Economic Times as 'Top Retention Solution for D2C'."
**Action:** "Book your 15-min strategy session before we're full → [Calendar Link]"

**Negative Constraints:** Avoid generic growth claims. Focus on technical implementation.
"""

# Prompt 2: COO of Pharma SME
coo_prompt = """
Generate a supply chain email for Pharma COOs using AIDCA + Reciprocity & Social Proof:

**Attention:** "60% of Pharma SMEs overpay due to inefficient logistics—sound familiar?"
**Interest:** "Peer companies reduced delays by 50% using our system (benchmarks inside)."
**Desire:** "Claim your free supply chain audit (normally $2000 value)."
**Conviction:** "92% of audited companies implement at least 3 cost-saving changes."
**Action:** "Reply 'AUDIT' to get your free workflow analysis."

**Negative Constraints:** No pharma jargon. Focus on operational outcomes.
"""

display(Markdown("### **Generated Email for D2C CTO**"))
display(Markdown(generate_email(cto_prompt)))

display(Markdown("### **Generated Email for Pharma COO**"))
display(Markdown(generate_email(coo_prompt)))
```

Prompt 2: Targeting COO of a Pharma SME (AIDCA + Cialdini Principles)

Goal: Email for COOs needing supply chain automation.


```python
# @title ### **Funnel Leak Detection & Repair**

# Simulated funnel data
data = {
    'Campaign': ['A', 'B', 'C'],
    'Leads': [1000, 800, 1200],
    'MQL (%)': [40, 60, 30],
    'SQL (%)': [10, 8, 25],
    'Client (%)': [2, 1, 5],
    'Drop-off Reason': ['Weak CTA', 'Over-engineered', 'Low authority'],
    'Message Summary': ['"Learn how we help"', '"AI hyperautomation"', '"Join 500+ brands"']
}

df = pd.DataFrame(data)

# Calculate absolute numbers
df['MQL'] = df['Leads'] * df['MQL (%)'] / 100
df['SQL'] = df['MQL'] * df['SQL (%)'] / 100
df['Clients'] = df['SQL'] * df['Client (%)'] / 100

# Funnel visualization
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
ax = sns.lineplot(data=df[['Leads', 'MQL', 'SQL', 'Clients']].T, dashes=False, markers=True)
ax.set_title("Funnel Performance by Campaign", pad=20, fontsize=16)
ax.set_ylabel("Count", fontsize=12)
ax.set_xlabel("Funnel Stage", fontsize=12)
plt.xticks(rotation=45)
plt.show()

# AI-powered diagnosis function
def diagnose_funnel(row):
    issues = []
    fixes = []

    if row['MQL (%)'] > 50 and row['SQL (%)'] < 15:
        issues.append("Interest→Desire gap")
        fixes.append("Add social proof + case studies")

    if row['SQL (%)'] > 20 and row['Client (%)'] < 5:
        issues.append("Trust deficit")
        fixes.append("Inject authority signals + testimonials")

    if row['MQL (%)'] < 35:
        issues.append("Weak hook")
        fixes.append("Reframe value prop with urgency")

    return pd.Series({
        'Main Issue': " | ".join(issues),
        'AI Recommended Fix': " | ".join(fixes)
    })

# Apply diagnosis
df = df.join(df.apply(diagnose_funnel, axis=1))

display(Markdown("### **Funnel Diagnostics Report**"))
display(df)
```

Part 3: Dashboard Design (Figma Wireframe Logic)

```python
# @title ### **Boardroom Decision Dashboard**

# Dashboard metrics calculation
dashboard_data = {
    'Metric': [
        'Overall Conversion Rate',
        'Best Performing Stage',
        'Worst Leak Point',
        'Top Campaign by ROI',
        'Recommended Intervention'
    ],
    'Value': [
        f"{df['Clients'].sum()/df['Leads'].sum()*100:.1f}%",
        "MQL Generation (Avg 43%)",
        "SQL→Client (Avg 2.7%)",
        "Campaign C (5% conversion)",
        "Trust-building at SQL stage"
    ],
    'Insight': [
        "3.2x below industry benchmark",
        "Strong targeting but weak nurturing",
        "Lack of case studies hurting conversions",
        "Higher authority signals perform best",
        "Implement testimonial-driven nurturing"
    ]
}

dashboard_df = pd.DataFrame(dashboard_data)

# Visual dashboard
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.barplot(x='Campaign', y='Client (%)', data=df)
plt.title("Conversion Rate by Campaign")

plt.subplot(2, 2, 2)
sns.heatmap(df[['MQL (%)', 'SQL (%)', 'Client (%)']], annot=True, cmap="YlGnBu")
plt.title("Funnel Heatmap")

plt.subplot(2, 2, 3)
df['ROI Score'] = df['Client (%)'] * df['Leads'] / 100
sns.scatterplot(x='Leads', y='Client (%)', size='ROI Score', data=df, sizes=(100, 500))
plt.title("Lead Volume vs Conversion Efficiency")

plt.tight_layout()
plt.show()

display(Markdown("### **Strategic Metrics Dashboard**"))
display(dashboard_df)
```
